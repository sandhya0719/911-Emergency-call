```python
dd['timeStamp']=pd.to_datetime(dd['timeStamp'])
type(dd['timeStamp'].iloc[0])
```

```python
time=dd['timeStamp'].iloc[0]
time.hour
```

```python
time.month
```

```python
time.dayofweek
```

```python
dd['Hour']=dd['timeStamp'].apply(lambda time:time.hour)
```

```python
dd['Hour']
```

```python
dd['Month']=dd['timeStamp'].apply(lambda time:time.month)
dd['Day of Week']=dd['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
dd['Day of Week']
```

```python
dmap={0:'Mon',1:'Tue',2:'Wen',3:'Thur',4:'fri'}
dd['Day of Week']=dd['Day of Week'].map(dmap)
```

```python
dd.head()
```

```python
sns.countplot(x='Day of Week',data=dd,hue='Reason')
```

```python
sns.countplot(x='Month',data=dd)
```

```python
bymonth=dd.groupby('Month').count()      # Allow to u together row based off of a column and perform an aggregate function on them
```

```python
bymonth
```

```python
bymonth.head()
```

```python
bymonth['lat'].plot()
```

```python
bymonth.reset_index()   #method sets a list of integer ranging from 0 to length of data as index
```

```python
t=dd['timeStamp'].iloc[0]
t
```

```python
dd['Date']=dd['timeStamp'].apply(lambda t:t.date())
```

```python
dd.head()
```

```python
dd.groupby('Date').count().head()
```

```python
dd.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
dd[dd['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
dd[dd['Reason']=='Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
dd[dd['Reason']=='EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
ddHours=dd.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
ddHours
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(ddHours,cmap='viridis')
```

```python
sns.clustermap(ddHours,cmap='viridis')
```

```python
ddMonth=dd.groupby(['Day of Week','Month']).count()['Reason'].unstack()
ddMonth
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(ddMonth,cmap='viridis')
```

```python
sns.clustermap(ddMonth,cmap='viridis')
```
